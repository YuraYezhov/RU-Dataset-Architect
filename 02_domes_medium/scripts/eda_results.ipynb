### Загрузка

```python
import pandas as pd
from pathlib import Path

BASE_DIR = Path.cwd().parent
DATASET_PATH = BASE_DIR / "data" / "domes_medium_dataset_raw.csv"

if DATASET_PATH.exists():
    df = pd.read_csv(DATASET_PATH)
    print("Датасет успешно загружен.")
    print(f"Строк: {len(df)}")
else:
    print(f"Файл не найден по пути: {DATASET_PATH}")    
```

### Анализ

```python
# Размерность
print(f"\nРазмер датасета (строк, колонок): {df.shape}")

# Первые 10 строк
print("\nПервые 10 строк:")
df.head(10)
```

```python
# Общая техническая информация
print("\nТехническая информация:")
df.info()
```

```python
# Проверка на пропуски (NaN/Null)
print("\nКоличество пустых значений в каждой колонке:")
null_counts = df.isnull().sum()
print(null_counts)
    
# Процент пустых строк
print(f"\nПроцент пустых данных: {(df.isnull().sum().sum() / df.size) * 100:.2f}%")
```

```python
# Статистическое описание (для числовых колонок)
print("\nОсновные статистики (Mean, Std, Min, Max):")
df.describe()
```

```python
# Распределение по категориям (если есть колонка century или target_shape)
if 'target_shape' in df.columns:
    print("\nРаспределение по формам (target_shape):")
    print(df['target_shape'].value_counts())

if 'century' in df.columns:
    print("\nРаспределение по векам (century):")
    print(df['century'].value_counts())
```
