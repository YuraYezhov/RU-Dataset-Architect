```python
import pandas as pd
from pathlib import Path

BASE_DIR = Path.cwd().parent
DATASET_PATH = BASE_DIR / "data" / "data-15-structure-3.csv"

if DATASET_PATH.exists():
    df_orig = pd.read_csv(DATASET_PATH)
    print("Датасет успешно загружен.")
    print(f"Строк: {len(df_orig)}")
else:
    print(f"Файл не найден по пути: {DATASET_PATH}")  
```

```python
# Размерность
print(f"Размер датасета: {df_orig.shape}")

# Первые 5 строк
print("\nПервые 5 строк")
df_orig.head(5)
```

```python
# Общая техническая информация
print("Техническая информация:")
df_orig.info()
```

```python
# Проверка на пропуски (NaN/Null)

# Считаем общее количество строк
total_rows = len(df_orig)

# Считаем процент пропусков для КАЖДОЙ колонки отдельно
null_percentages = (df_orig[['Высота', 'Ширина', 'Длина', 'Вес']].isnull().sum() / total_rows) * 100

print("Процент пропусков по каждой колонке:")
print(null_percentages.map('{:.2f}%'.format))

# Средний процент "пустоты" именно по этим 4-м колонкам
total_nulls_in_subset = df_orig[['Высота', 'Ширина', 'Длина', 'Вес']].isnull().sum().sum()
total_cells_in_subset = total_rows * 4
actual_empty_percent = (total_nulls_in_subset / total_cells_in_subset) * 100

print(f"Средняя 'пустота' ключевых признаков: {actual_empty_percent:.2f}%")
```

```python
print("Распределение всех категорий в исходных данных:")
category_counts = df_orig['Классификация культурной ценности'].value_counts()
print(category_counts)
```
