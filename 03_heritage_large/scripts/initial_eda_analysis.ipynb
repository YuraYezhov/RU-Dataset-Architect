#### Приведение к одинаковым единицам измерения

```python
import numpy as np
import re

# Словари от ИИ-агента
TO_CM = {
    'см': 1.0, 'см.': 1.0, 'cm': 1.0, 'cм': 1.0, 'cм.': 1.0,
    'мм': 0.1, 'мм.': 0.1, 'mm': 0.1,
    'м': 100.0, 'м.': 100.0, 'm': 100.0,
    'дм': 10.0, 'дм.': 10.0, 'dm': 10.0,
    'дюйм': 2.54,  # английский; если французский — 2.707
    'вершок': 4.445, 'верш': 4.445,
    'аршин': 71.12, 'арш': 71.12,
    'фут': 30.48, 'ft': 30.48,
}
TO_GRAM = {
    'г': 1.0, 'г.': 1.0, 'гр': 1.0, 'гр.': 1.0, 'g': 1.0, 'gr': 1.0,
    'кг': 1000.0, 'кг.': 1000.0, 'kg': 1000.0,
    'мг': 0.001, 'мг.': 0.001, 'mg': 0.001,  # миллиграммы!
}
```

```python
def smart_convert(value, is_weight=False):
    """
    
    """
    # Превращаем в строку и очищаем от пробелов
    s = str(value).lower().strip()

    # Если ячейка пустая (была NaN или стала ""), возвращаем честный NaN
    if s in ['nan', 'none', '']:
        return np.nan
    
    # Извлекаем первое число из строки
    match = re.search(r"(\d+\.?\d*)", s)
    if not match:
        return np.nan
    
    try:
        num = float(match.group(1).replace(',', '.'))

        # Ищем единицу измерения
        multiplier = None
        if is_weight:
            lookup = TO_GRAM
        else:
            lookup = TO_CM
        
        # Ищем любую единицу из словаря в строке
        for unit, val in lookup.items():
            if unit in s:
                multiplier = val
                break
        
        # Если единиц нет (просто число) — берем значение по умолчанию
        if multiplier is None:
            return num # По умолчанию см или г
        
        return num * multiplier
    
    except:
        return np.nan
```

```python
def extract_weight_from_text(row):
    """
    
    """
    if pd.notna(row['weight']): 
        return row['weight']
    
    text = f"{row['title']} {row['marks']}".lower()
    # Ищем паттерн: "вес 720 гр" или "вес: 8,5 кг"
    match = re.search(r"вес\s*:?\s*(\d+[.,]?\d*)\s*(гр|г|кг|kg|g)", text)
    if match:
        val = float(match.group(1).replace(',', '.'))
        unit = match.group(2)
        return val * TO_GRAM.get(unit, 1.0)
    
    return np.nan
```

```python
# Сначала чистим то, что уже в колонках
for col in ['height', 'width', 'length']:
    df_raw[col] = df_raw[col].apply(smart_convert, is_weight=False)

df_raw['weight'] = df_raw['weight'].apply(smart_convert, is_weight=True)

# 2. Дозаполняем ВЕС из текстовых описаний
df_raw['weight'] = df_raw.apply(extract_weight_from_text, axis=1)

# Проверка результата
print("Умная конвертация завершена!")
print("Статистика очистки:")
for col in ['height', 'width', 'length', 'weight']:
    success_count = df_raw[col].notna().sum()
    print(f"Колонка {col}: успешно преобразовано {success_count} числовых значений")
```

#### Заполнение пустых ячеек

```python
RE_3D = re.compile(r'(\d+[.,]?\d*)\s*[xх*×]\s*(\d+[.,]?\d*)\s*[xх*×]\s*(\d+[.,]?\d*)', re.IGNORECASE)
RE_WEIGHT = re.compile(r'вес\s*:?\s*(\d+[.,]?\d*)\s*(гр|г|кг|kg|g|мг)', re.IGNORECASE)

# Словари конвертации
TO_G = {'г': 1.0, 'гр': 1.0, 'кг': 1000.0, 'мг': 0.001}
```

```python
def extract_missing_info(row):
    """Пытается вытащить Длину и Вес из текста, если их нет в колонках."""
    text = (str(row['title']) + " " + str(row['marks'])).lower()

    # Извлекаем Длину (если пустая) из паттерна 10х20х30
    if pd.isna(row['length']):
        match_3d = RE_3D.search(text)
        if match_3d:
            row['length'] = float(match_3d.group(3).replace(',', '.'))
            # Если высота или ширина были пустыми — дозаполняем их из этого же паттерна
            if pd.isna(row['height']): row['height'] = float(match_3d.group(1).replace(',', '.'))
            if pd.isna(row['width']): row['width'] = float(match_3d.group(2).replace(',', '.'))

    # Извлекаем Вес (если пустой)
    if pd.isna(row['weight']):
        match_w = RE_WEIGHT.search(text)
        if match_w:
            val = float(match_w.group(1).replace(',', '.'))
            unit = match_w.group(2)
            row['weight'] = val * TO_G.get(unit, 1.0)
            
    return row
```

```python
# --- ШАГ 1: Восстановление данных из текста ---
df_raw = df_raw.apply(extract_missing_info, axis=1)

# --- ШАГ 2: Заполнение медианой (STANDALONE STEP) ---
# Заполняем высоту и ширину медианой ПО ГРУППАМ
df_raw['height'] = df_raw['height'].fillna(df_raw.groupby('target_group')['height'].transform('median'))
df_raw['width'] = df_raw['width'].fillna(df_raw.groupby('target_group')['width'].transform('median'))

# --- ШАГ 3: Финальная обработка Длины и Веса ---
# Длину и вес медианой заполнять опасно (слишком много пустоты), 
# поэтому заменяем оставшиеся NaN на 0, чтобы ML-модель могла работать.
df_raw['length'] = df_raw['length'].fillna(0)
df_raw['weight'] = df_raw['weight'].fillna(0)

# Проверка результатов
print("Пропусков после обработки:")
print(df_raw[['height', 'width', 'length', 'weight']].isna().sum())
```

```python
import re
# оригинал самый первый
def fill_missing_dimensions(df):
    """

    """
    initial_missing = df[['height', 'width']].isna().sum().sum()

    # Регулярные выражения
    # 1. Размер "Число x Число" (латинская 'x', кириллическая 'х', '*')
    dim_pattern = re.compile(r'(\d+(?:[.,]\d+)?)\s*[xх*]\s*(\d+(?:[.,]\d+)?)', re.IGNORECASE)
    # 2. "диаметр Число"
    diam_pattern = re.compile(r'(?:диаметр|диам\.)\s*(\d+(?:[.,]\d+)?)', re.IGNORECASE)

    def extract_dims(row):
        # Если оба уже есть, ничего не делаем
        if pd.notna(row['height']) and pd.notna(row['width']):
            return row['height'], row['width']

        text = str(row['title']) + " " + str(row['marks'])

        # Поиск 2D размеров
        match_dim = dim_pattern.search(text)
        if match_dim:
            h = float(match_dim.group(1).replace(',', '.'))
            w = float(match_dim.group(2).replace(',', '.'))
            return h, w

        # Поиск диаметра
        match_diam = diam_pattern.search(text)
        if match_diam:
            val = float(match_diam.group(1).replace(',', '.'))
            return val, val

        return row['height'], row['width']

    # Применяем функцию
    new_dims = df.apply(extract_dims, axis=1, result_type='expand')
    df['height'] = new_dims[0]
    df['width'] = new_dims[1]

    final_missing = df[['height', 'width']].isna().sum().sum()
    print(f"Заполнено пропущенных значений: {initial_missing - final_missing}")
    return df

# Применяем к датафрейму
df_raw = fill_missing_dimensions(df_raw)
```
